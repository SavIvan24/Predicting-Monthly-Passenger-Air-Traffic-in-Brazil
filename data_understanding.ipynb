# Data Understanding

## 1. Collect initial data

Data source: https://www.kaggle.com/datasets/sturarods/anac-national-civil-aviation-agency-2000-2025?select=anac_brazil.csv

Selection criteria:
- file of interest: `anac_brazil.csv`
- necessary attributes: temporal attributes (date of a flight), target variables (number of passengers, cargo weight), filtering attributes (type of a flight, country of departure/arrival for identifying domestic/international flight).
- irrelevant attribures: airline and aicraft details, redundant date components (day, time, week, semester), baggage metrics, file metadata.

## 2. Describe data

It is calculated, in each basic step, by the product between the quantity, expressed in kilograms, of passengers, luggage, cargo and mail paid transported and the distance of the stage, divided by 1000.

### Attribute analysis

1. Attribute correlations:
    - *Time correlation:* 14-24 columns encode departure reference date, 33-44 columns encode departure real date, 56-67 columns encode arrival real date (for each event about 11 columns encode only one date just at different level of detalization).
    - *Distance and capacity:* `km_distancia` and `nr_ask`, `nr_rpk`, `nr_atk`, `nr_rtk` - each of these attributes scales with distance; `kg_pesoatis`, `kg_correio`, `kg_bagagem_livre`, `kg_bagagem_excesso`, `nr_passag_pagos`, `nr_passag_gratis` - total aircraft weight is built from those attributes.
    - *Passangers and load:* `nr_passag_pagos` and `nr_rpk`, `nr_passag_gratis` and `nr_pax_gratis_km`, `kg_carga_paga` and `nr_carga_paga_km`, `kg_carga_gratis` and `nr_carga_gratis_km`, `kg_correio` and `nr_correio_km`, `kg_bagagem_livre` and `nr_bagagem_gratis_km`, `kg_bagagem_excesso` and `nr_bagagem_paga_km` - these pairs of attributes are mathematically linked, since each `*_km` variable is the product of a load/passenger measure and distance.

2. Basic statistics analysis:
    - Passenger volume per flight:
      - The median of total passengers is 105. With an error this value is typical for narrow‑body jets designed for short- and medium-range routes (domestic flights).
      - Non‑revenue passengers are negligible on average.
      - Zero as min value of revenue passengers indicates non-regular flights (can be training or cancelled flights). This must be filtered with `ds_grupo_di` = 'REGULAR' to avoid artificial zeros in monthly time series.

    - Cargo weight per flight:
      - Distribution has right–sided asymmetry - mean of 10 360 kg is much higher than median 50 kg. It means that most flights carry very little paid cargo and small number of flights carry a lot of paid cargo.
      - The 25th percentile is 0, which means that quarter of all records do not contain information about cargo weight. It can be input error.
      - Values that are bigger than 130 000 - 140 000 kg are likely an input error (carrying capacity of any aircraft that regularly flies to Brazil does not allow it to carry 150 000 - 230 000 kg in one flight).

    - Temporal attributes:
      - The 25th percentile date is the end of 2007 which means that the first seven years have sparser data.
      
    - Network breadth:
      - Diversity of airports indicates big number of small private airports with non stable schedule and passenger traffic. Predicting monthly traffic at national level is the most appropriate option rather than predicting monthly traffic at each airport.
      - Brazil has high international connectivity (>150 contries), so most likely the traffic was impacted by COVID.

    - Flight category:
      - Target population for forecasting is `ds_grupo_di` = 'REGULAR' (excluding charters and non‑commercial flights).
      - Both `ds_natureza_tipo_linha` and `ds_servico_tipo_linha` having 1942493 not identified records seems to be an error when writing data. Values of `ds_natureza_tipo_linha` and `ds_servico_tipo_linha` can be derived from other attributes (`nm_pais_origem`, `nm_pais_destino`, `nr_passag_pagos`).

Data balancing is not necessary because our main task is time-series prediction, not classification.

### Basic analysis to verify the hypothesis



1. Passenger traffic follows annual seasonality.

```python
from statsmodels.tsa.seasonal import seasonal_decompose

chunk_size = 100000
chunks = []

reader = pd.read_csv(
    'anac_brazil.csv',
    usecols=['nr_ano_mes_partida_real', 'nr_passag_pagos', 'ds_grupo_di'],
    chunksize=chunk_size
)

for chunk in reader:
    chunk = chunk[chunk['ds_grupo_di'] == 'REGULAR']
    chunk = chunk.dropna(subset=['nr_passag_pagos', 'nr_ano_mes_partida_real'])
    monthly = chunk.groupby('nr_ano_mes_partida_real')['nr_passag_pagos'].sum().reset_index()
    chunks.append(monthly)
    
monthly_df = pd.concat(chunks).groupby('nr_ano_mes_partida_real').sum().reset_index()
```

```python
monthly_df = monthly_df.sort_values('nr_ano_mes_partida_real')
monthly_df['date'] = pd.to_datetime(monthly_df['nr_ano_mes_partida_real'], format='%Y%m')
monthly_df.set_index('date', inplace=True)

plt.figure(figsize=(12, 5))
plt.plot(monthly_df.index, monthly_df['nr_passag_pagos'], color='navy')
plt.title('Monthly revenue passengers')
plt.ylabel('Passengers')
plt.grid(alpha=0.3)
plt.show()

full_index = pd.date_range(start=monthly_df.index.min(), end=monthly_df.index.max(), freq='MS')
monthly_df = monthly_df.reindex(full_index)
decomp = seasonal_decompose(monthly_df['nr_passag_pagos'], model='additive', period=12)
decomp.plot()
plt.suptitle('Seasonal passenger traffic', y=1.02)
plt.tight_layout()
plt.show()
```

2. COVID pandemic influenced international traffic more than domestic.

```python
chunk_size = 100000
chunks = []

reader = pd.read_csv(
    'anac_brazil.csv',
    usecols=['nr_ano_mes_partida_real', 'nr_passag_pagos', 'ds_grupo_di', 'ds_natureza_tipo_linha'],
    chunksize=chunk_size
)

for chunk in reader:
    chunk = chunk[chunk['ds_grupo_di'] == 'REGULAR']
    chunk = chunk[chunk['ds_natureza_tipo_linha'].isin(['DOMÉSTICA', 'INTERNACIONAL'])]
    chunk = chunk.dropna(subset=['nr_passag_pagos'])
    monthly = chunk.groupby(['nr_ano_mes_partida_real', 'ds_natureza_tipo_linha'])['nr_passag_pagos'].sum().reset_index()
    chunks.append(monthly)

nature_df = pd.concat(chunks).groupby(['nr_ano_mes_partida_real', 'ds_natureza_tipo_linha']).sum().reset_index()
```

```python
nature_df = nature_df.sort_values('nr_ano_mes_partida_real')
nature_df['date'] = pd.to_datetime(nature_df['nr_ano_mes_partida_real'], format='%Y%m')
nature_df.set_index('date', inplace=True)

domestic_df = nature_df[nature_df['ds_natureza_tipo_linha'] == 'DOMÉSTICA']
international_df = nature_df[nature_df['ds_natureza_tipo_linha'] == 'INTERNACIONAL']

plt.figure(figsize=(12, 5))
plt.plot(domestic_df.index, domestic_df['nr_passag_pagos'], color='navy')
plt.title('Monthly revenue passengers on domestic flights')
plt.ylabel('Passengers')
plt.grid(alpha=0.3)
plt.show()

plt.figure(figsize=(12, 5))
plt.plot(international_df.index, international_df['nr_passag_pagos'], color='navy')
plt.title('Monthly revenue passengers on international flights')
plt.ylabel('Passengers')
plt.grid(alpha=0.3)
plt.show()
```

4. There is a correlation between the monthly volumes of passengers and cargo traffic.

```python
from scipy.stats import pearsonr, spearmanr
from statsmodels.tsa.stattools import ccf

chunk_size = 100000
chunks = []

reader = pd.read_csv(
    'anac_brazil.csv',
    usecols=[
        'ds_grupo_di', 'ds_natureza_tipo_linha', 'nm_pais_origem',
        'nm_pais_destino', 'ds_servico_tipo_linha',
        'nr_passag_pagos', 'kg_carga_paga', 'nr_ano_mes_partida_real'
    ],
    chunksize=chunk_size
)

for chunk in reader:
    chunk = chunk[chunk['ds_grupo_di'] == 'REGULAR']
    chunk = chunk[chunk['kg_carga_paga'] <= 140000]
    chunks.append(chunk)

cargo_pass_correlation = pd.concat(chunks, ignore_index=True)
```

```python
# group passengers and cargo by months 
def monthly_aggregate(data, segment_label):
    monthly = data.groupby('nr_ano_mes_partida_real').agg(
        passengers=('nr_passag_pagos', 'sum'),
        cargo_kg=('kg_carga_paga', 'sum')
    ).reset_index()
    monthly['year_month'] = pd.to_datetime(monthly['nr_ano_mes_partida_real'], format='%Y%m')
    monthly = monthly.sort_values('year_month').reset_index(drop=True)
    monthly['segment'] = segment_label
    return monthly

domestic = cargo_pass_correlation[cargo_pass_correlation['ds_natureza_tipo_linha'] == 'DOMÉSTICA']
international = cargo_pass_correlation[cargo_pass_correlation['ds_natureza_tipo_linha'] == 'INTERNACIONAL']

dom_monthly = monthly_aggregate(domestic, 'DOMESTIC')
int_monthly = monthly_aggregate(international, 'INTERNATIONAL')
```

```python
fig, axes = plt.subplots(2, 1, figsize=(14, 8), sharex=True)

for ax, data, title in zip(axes, [dom_monthly, int_monthly],
                            ['Domestic', 'International']):
    ax1 = ax
    ax1.plot(data['year_month'], data['passengers']/1e6, color='navy', label='Passengers (mln)')
    ax1.set_ylabel('Passengers (millions)', color='navy')
    ax2 = ax1.twinx()
    ax2.plot(data['year_month'], data['cargo_kg']/1e6, color='darkorange', label='Cargo (kton)')
    ax2.set_ylabel('Cargo (thousands of tons)', color='darkorange')
    ax.set_title(f'{title} Monthly Passengers vs Cargo')
    ax1.legend(loc='upper left')
    ax2.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

### Inference

1. Seasonality of revenue passenger traffic is consistent and predictable. On average the number of revenue passengers grows every year excluding expected drops in 2016 (economic, political and social crisis in Brazil) and 2020 (COVID-19 pandemic).
2. After COVID-19 pandemic international traffic was recovering much slower than domestic. Modeling domestic and international flights separately will improve forecast accuracy.
3. After ~2006-2007 the ration of non-revenue/revenue passengers remained the same - ~0.02. Predicting only revenue passengers can be valid simplification.
4. After 2011-2012 number of passengers and cargo weight had strong correlation (both domestic and international) but it dissapeared after COVID-19 for international flights while domestic flights still have some correlation. It seems better to predict cargo weight and number of passengers separately.
