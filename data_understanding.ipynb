### Columns description

1. **id_basica**: Numeric code identifier of the basic step in the system.
2. **id_empresa**: no description
3. **sg_empresa_icao**: ICAO acronym of the airline. It refers to the designator of the air transport company obtained from the ICAO (International Civil Aviation Organization).
4. **sg_empresa_iata**: IATA acronym of the airline. Refers to the designator of the air transport company obtained from IATA (International Air Transport Association).
5. **nm_empresa**: Name of the airline.
6. **nm_pais**: Name of the country of the nationality of the airline.
7. **ds_tipo_empresa**: Description of the type of company. It refers to the description of the type of company in relation to the service performed. It may contain the following descriptions. This variable is in disuse.
8. **nr_voo**: no description
9. **nr_singular**: Flight Singularity. It refers to the set of characters that helps in the identification of the flight, composed of letters and numbers chosen at the discretion of the airline itself.
10. **id_di**: Numerical code that references the identifier digit variable in the system.
11. **cd_di**: Identifier Digit. It refers to the character used to identify the type of authorization for each flight stage.
12. **ds_di**: Description DI. Presents the description corresponding to the Identifier Digit.
13. **ds_grupo_di**: Description of the DI group or Flight Type. Refers to the grouping of DI classifications into types of operations.
14. **dt_referencia**: Reference Date. Corresponds to the Expected Start Date of the Flight.
15. **nr_ano_referencia**: Reference Year.
16. **nr_semestre_referencia**: Reference semester number.
17. **nm_semestre_referencia**: Name of the reference semester.
18. **nr_trimestre_referencia**: Reference quarter number.
19. **nm_trimestre_referencia**: Name of the reference quarter.
20. **nr_mes_referencia**: Reference month number.
21. **nm_mes_referencia**: Name of the reference month.
22. **nr_semana_referencia**: Reference week number.
23. **nm_dia_semana_referencia**: Name of the reference week day.
24. **nr_dia_referencia**: Reference day number.
25. **nr_ano_mes_referencia**: Number representing the year and month of reference, in the format YYYYMM.
26. **id_tipo_linha**: Numerical code identifier of the line type in the system.
27. **cd_tipo_linha**: Character that identifies the type of line.
28. **ds_tipo_linha**: Description of the line type.
29. **ds_natureza_tipo_linha**: Description of the nature referring to the type of line (International/Domestic). Consider the nature of the flight, not just the stage.
30. **ds_servico_tipo_linha**: Description of the service referring to the type of line (Passenger/Cargo).
31. **ds_natureza_etapa**: Description of the nature of the flight stage (International/Domestic).
32. **hr_partida_real**: Flight departure time, calculated by the criterion of the calço and descalço.
33. **dt_partida_real**: Date of realization. It refers to the full take-off date of the stage, calculated by the criterion of the shoe and barefoot.
34. **nr_ano_partida_real**: Number of the actual departure year.
35. **nr_semestre_partida_real**: Real departure semester number.
36. **nm_semestre_partida_real**: Name of the real departure semester.
37. **nr_trimestre_partida_real**: Number of the actual departure quarter.
38. **nm_trimestre_partida_real**: Name of the actual departure quarter.
39. **nr_mes_partida_real**: Number of the real departure month.
40. **nm_mes_partida_real**: Name of the real departure month.
41. **nr_semana_partida_real**: Number of the actual departure week.
42. **nm_dia_semana_partida_real**: Name of the day of the real departure week.
43. **nr_dia_partida_real**: Number of the actual departure day.
44. **nr_ano_mes_partida_real**: Number that represents the year and the month of actual departure, in the format YYYYMM.
45. **id_aerodromo_origem**: Identifier code of the airfield of origin in the system.
46. **sg_icao_origem**: ICAO acronym of the aerodrome of origin.
47. **sg_iata_origem**: IATA acronym of the airfield of origin.
48. **nm_aerodromo_origem**: Name of the aerodrome of origin.
49. **nm_municipio_origem**: Name of the municipality of origin.
50. **sg_uf_origem**: Acronym of the Federative Unit of origin.
51. **nm_regiao_origem**: Name of the region of origin.
52. **nm_pais_origem**: Name of the country of origin.
53. **nm_continente_origem**: Name of the continent of origin.
54. **nr_etapa**: Step number. Indicates the sequence of the aerodrome of origin between the aerodromes under the same flight number.
55. **hr_chegada_real**: Flight arrival time, determined by the criteria of the calço and descalço.
56. **dt_chegada_real**: Actual date of arrival.
57. **nr_ano_chegada_real**: Number of the year of actual arrival.
58. **nr_semestre_chegada_real**: Real arrival semester number.
59. **nm_semestre_chegada_real**: Name of the real arrival semester.
60. **nr_trimestre_chegada_real**: Number of the actual arrival quarter.
61. **nm_trimestre_chegada_real**: Name of the real arrival quarter.
62. **nr_mes_chegada_real**: Number of the real month of arrival.
63. **nm_mes_chegada_real**: Name of the real month of arrival.
64. **nr_semana_chegada_real**: Number of the actual arrival week.
65. **nm_dia_semana_chegada_real**: Name of the day of the week of actual arrival.
66. **nr_dia_chegada_real**: Number of the actual day of arrival.
67. **nr_ano_mes_chegada_real**: Number that represents the year and month of actual arrival, in the format YYYYMM.
68. **id_equipamento**: Identification code of the aircraft model in the system.
69. **sg_equipamento_icao**: ICAO Designator of the aircraft model ("Type Designator").
70. **ds_modelo**: Description of the aircraft model.
71. **ds_matricula**: Aircraft registration mark.
72. **id_aerodromo_destino**: Identification code of the destination aerodrome in the system.
73. **sg_icao_destino**: ICAO acronym of the destination airfield.
74. **sg_iata_destino**: IATA acronym of the destination aerodrome.
75. **nm_aerodromo_destino**: Name of the destination aerodrome.
76. **nm_municipio_destino**: Name of the destination municipality.
77. **sg_uf_destino**: Acronym of the destination Federative Unit.
78. **nm_regiao_destino**: Name of the destination region.
79. **nm_pais_destino**: Name of the destination country.
80. **nm_continente_destino**: Name of the destination continent.
81. **nr_escala_destino**: Destination scale sequence. It is the number corresponding to the destination aerodrome of the stage, considering the sequential numbering of the aerodromes traveled under the same flight.
82. **lt_combustivel**: Amount of fuel consumed, expressed in liters.
83. **nr_assentos_ofertados**: Number of seats available at the flight stage.
84. **kg_payload**: Payload capacity, expressed in kilograms.
85. **km_distancia**: Distance, expressed in kilometers, between the aerodromes of origin and destination of the stage, considering the curvature of planet earth.
86. **nr_passag_pagos**: Number of passengers who occupy seats marketed to the public and who generate revenue, with the purchase of seats, for the air transport company.
87. **nr_passag_gratis**: Number of passengers who occupy seats marketed to the public, but do not generate revenue, with the purchase of seats, for the air transport company.
88. **kg_bagagem_livre**: Franchised baggage. It is all baggage that is not considered as excess, expressed in kilograms.
89. **kg_bagagem_excesso**: Excess baggage. It is the total amount of baggage that exceeds the weight limit agreed between the airline and the passenger, checked at the time of dispatch, expressed in kilograms.
90. **kg_carga_paga**: Cargo paid. It is the total amount, expressed in kilograms, of all goods that have been transported on the aircraft, except mail and luggage, and have generated direct or indirect revenue for the airline.
91. **kg_carga_gratis**: Total quantity, expressed in kilograms, of all goods that have been transported on the aircraft, except mail and baggage, and have not generated direct or indirect revenues for the airline.
92. **kg_correio**: Quantity, expressed in kilograms, of objects transported to serve the operators officially designated by the country to operate postal services and comply with the associated obligations arising from the Acts of the Universal Postal Union (UPU).
93. **nr_decolagem**: Number of takeoffs.
94. **nr_horas_voadas**: Number of hours flown.
95. **kg_peso**: Refers to the total weight loaded by the aircraft, calculated by the sum of cargo, mail, passengers and luggage, expressed in kg.
96. **nr_velocidade_media**: Average flight speed, calculated from the distance and duration of the flight, expressed in km/h.
97. **nr_pax_gratis_km**: Represents, for each basic step, the product between the number of free passengers and the distance of the stage.
98. **nr_carga_paga_km**: Refers to the product between the amount of load paid (kg) and the distance of the basic step (km).
99. **nr_carga_gratis_km**: Refers to the product between the amount of free load (kg) and the distance of the basic step (km).
100. **nr_correio_km**: Refers to the product between the amount of mail (kg) and the distance of the basic step (km).
101. **nr_bagagem_paga_km**: Refers to the product between the amount of baggage paid (kg) and the distance of the basic step (km).
102. **nr_bagagem_gratis_km**: Refers to the product between the amount of free baggage (kg) and the distance from the basic step (km).
103. **nr_ask**: Refers to the volume of Seat Kilometers Offered (Available Seat Kilometer). It is calculated, in each basic step, by the product between the number of seats offered and the distance of the stage.
104. **nr_rpk**: Refers to the volume of Passenger Kilometers Transported (Revenue Passenger Kilometer). It is calculated, in each basic step, by the product between the number of passengers paid and the distance of the stage.
105. **nr_atk**: Refers to the volume of Tons Kilometers Offered (Available Tonne Kilometer), expressed in (ton x km). It is calculated, in each basic step, by the product between the Payload Capacity (kg) and the distance of the step, divided by 1000.
106. **nr_rtk**: Refers to the volume of Tonnes Transported Kilometers (Revenue Tonne Kilometer), expressed in (ton x km). It is calculated, in each basic step, by the product between the quantity, expressed in kilograms, of passengers, luggage, cargo and mail paid transported and the distance of the stage, divided by 1000.
107. **id_arquivo**: Identifier code of the data file sent to the system.
108. **nm_arquivo**: File name sent to the system.
109. **nr_linha**: File line number in which the data of each basic step are contained.
110. **dt_sistema**: Date on which the information was recorded in the system.
111. **nr_chave**: no description

### Attribute analysis

1. Attribute correlations:
    - *Time correlation:* `dt_referencia`, `nr_ano_referencia`, `nr_mes_referencia`, `nr_dia_referencia`, `nr_ano_mes_referencia`, `nr_semana_referencia`, `nr_trimestre_referencia` - these attributes encode the same date/time at different level of detalization.
    - *Distance and capacity:* `km_distancia` and `nr_ask`, `nr_rpk`, `nr_atk`, `nr_rtk` - each of these attributes scales with distance; `kg_pesoatis`, `kg_correio`, `kg_bagagem_livre`, `kg_bagagem_excesso`, `nr_passag_pagos`, `nr_passag_gratis` - total aircraft weight is built from those attributes.
    - *Passangers and load:* `nr_passag_pagos` and `nr_rpk`, `nr_passag_gratis` and `nr_pax_gratis_km`, `kg_carga_paga` and `nr_carga_paga_km`, `kg_carga_gratis` and `nr_carga_gratis_km`, `kg_correio` and `nr_correio_km`, `kg_bagagem_livre` and `nr_bagagem_gratis_km`, `kg_bagagem_excesso` and `nr_bagagem_paga_km` - these pairs of attributes are mathematically linked, since each `*_km` variable is the product of a load/passenger measure and distance.
2. Basic statistics analysis and its meaning in business terms:

The dataset contains highly heterogeneous flights. For example, `km_distancia` has standard deviation of 1800 km, which means the data includes both domestic and international flights and therefore reflects different operational contexts.

Passenger-related variables are relatively stable compared to cargo variables: `nr_passag_pagos` is centered around 100 passengers per leg, while `nr_passag_gratis` is much lower, which suggests that most traffic comes from revenue passengers. In contrast, cargo attributes such as `kg_carga_paga` and `kg_correio` are highly skewed, with many zero or small values and a few very large ones, showing that cargo transport is concentrated in a smaller number of flights.

This means that business decisions based on the dataset should not rely only on averages, because the averages hide strong differences between route types, airlines, and flight purposes.
